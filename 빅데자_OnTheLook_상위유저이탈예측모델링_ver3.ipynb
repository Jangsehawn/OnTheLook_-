## LGBM 분류

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score, precision_score
from lightgbm import plot_importance
import matplotlib.pyplot as plt
import re
```

```python
# lgbm이 인식하도록 카테고리로 변경
#  high cardinality categorical features 일떄 원핫인코딩말고 
# lgbm은 범주형 변수를 2개의 subset으로 잘 나눔, 클래스를 분류하는데 가장 이득을 많이 줄 수 있는 방법으로 이분화를 시킨다는 는 의미이다. 이렇게 되면 속도도 빨라지고 성능도 더 좋아진다고 한다.


for i in total.columns:
    total[i]=total[i].astype('category')
```

```python
# 컬럼명에 특수문자가 포함되면 lgbm이 인식못하므로 변경

total=total.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))

```

```python
total
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(total.drop(['event'],axis=1), total['event'], test_size=0.3, random_state=123)
```

```python
lgbm= LGBMClassifier(n_estimator=400)
```

```python
lgbm.fit(X_train,Y_train)
```

```python
preds=lgbm.predict(X_test)
```

## 언더샘플링은 해당 데이터셋에서 큰 효과가 없는 것으로 판명되었다

-> 1:1을 맞춰주면 성능이 조금 올라가지만 표본이 작아서 큰 의미는 없는 것으로 판명

1:2 정도로 샘플링하면 기존 결과와 크게 다르지 않은 모습을 보임

```python
from sklearn.metrics import precision_recall_curve
from matplotlib import pyplot
```

```python
# calculate model precision-recall curve
precision, recall,_= precision_recall_curve(Y_test,preds)
# plot the model precision-recall curve
pyplot.plot(recall, precision, marker='.', label='lgbm')
# axis labels
pyplot.xlabel('Recall')
pyplot.ylabel('Precision')
# show the legend
pyplot.legend()
# show the plot
pyplot.show()
```

 high dimensional한 상황에서 imbalanced 해결을 위해서 smote나 adaysn같은 방법은 추천 X
 
->거리 기반의 샘플링이기 때문에 고차원에서 단순히 거리에 의존한 방법이 잘 작동하지 않는 경우가 많기 때문

weighted cross entropy나 cost-sensitive learning, 아니면 차라리 undersampling을 추천 O 
비율이 1:1이 아닌 상황이라고 imbalanced 문제가 생기는 것도 아니므로 일단은 원 데이터로 분류기를 훈련해보는 것을 추천


cost-sensitive learning은 class_weight 옵션을 분류기에서 지원하는 경우->costcla


- SMOTE나 ADASYN과 같은 오버샘플링 방법은 추천드리지 않습니다. 유클리디안 기반이기 때문에 high-dimensional feature에 대해서는 오히려 결정 경계를 찾는 것을 방해합니다.
- 차라리 언더샘플링이 좋을 수 있지만 언급하신 것처럼 샘플 수가 확 줄어드는 경우가 있습니다.
- weighted cross entropy loss, focal loss와 같이 loss function을 변경해 주시거나, classfier에 class_weight와 같은 인자가 있다면 해당 인자에 적절한 값을 주셔서 cost-sensitive learning을 하시는 것을 추천드립니다. (관련되어 costcla, imblearn 패키지를 참고해보세요)

- 이진 분류 상황을 가정했을 때, 분류기의 역할은 0과 1 사이의 continuous output을 출력하는 것이고, 이 output을 바탕으로 0 또는 1이라는 decision을 하는 것은 엄연히 learning의 영역과는 별개입니다.
- 대부분의 분류기 알고리즘은 0.5를 바탕으로 0 또는 1이라는 라벨을 정합니다. 
- 문제 상황에 맞게 label을 정하기 위해 optimal cutoff value를 정해주는 방법이 있습니다. https://machinelearningmastery.com/threshold-moving-for-imbalanced-classification/

또 imbalance한 상황에서는 confusion matrix, accuracy, precision, recall, F1이 그다지 좋은 평가 메트릭이 아닐 수 있습니다. label이 아닌 probabilistic output으로 평가하는 Brier score, Log-loss 그 자체, AUROC, AUPRC와 같은 metric을 추천 드립니다!
